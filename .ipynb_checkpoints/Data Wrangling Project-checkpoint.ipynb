# Personality Scores Dataset

```python
pscores_df = pd.read_csv('personality_scores.csv', delimiter=';')
pscores_df.head()
```

```python
pscores_df.isnull().sum()

for score in pscores_df.isnull():
    if "Unnamed" in score:
        del pscores_df[score]
        
pscores_df
```

```python
assert len(pscores_df['ID'].duplicated()) == len(pscores_df.ID.unique())
```

```python
pscores_df = pscores_df.set_index('ID')
pscores_df.head()
```

```python
pscores_df.columns
```

```python
import re
```

```python
cols = []

for col in pscores_df.columns:
    if '[' in col:
            
        first_bracket = col.index('[') 
        last_bracket = col.index(']')
        new_col = col[first_bracket+1:last_bracket]
        cols.append(new_col)
    else:
        cols.append(col)
    
pscores_df.columns = cols
pscores_df.head()
```

# Calculating Personality Scores

```python
catogories = {
    1: {
        "name": "Extraversion",
        "score": 0
       },
    2: {
        "name":"Agreeableness",
        "score": 0
        },
    3: {
        "name":"Conscientiousness",
        "score": 0
    },
    4: {
        "name":"Neuroticism",
        "score": 0
    },
    5: {
        "name":"Intellect/Imagination",
        "score": 0
    }
}

def sum_personality(score):
    
    if type(score[0]) == str:
        score[0] = tuple(map(int, score[0].strip('()').split(',')))
    
    for cat in range(1,6):

        print(cat)
        if cat == score[0][0]:
            catogories[cat]["score"] += score[0][1]
```

```python
sum_personality(dpt_pscores['I am always prepared.'])
catogories
# type(dpt_pscores['I am always prepared.'][0])
```
