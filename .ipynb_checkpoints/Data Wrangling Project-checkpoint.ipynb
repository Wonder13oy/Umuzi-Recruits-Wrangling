# Calculating Personality Scores

```python


catogories = {
        1: {
            "name": "Extraversion",
            "code": 1
           },
        2: {
            "name":"Agreeableness",
            "code": 2
            },
        3: {
            "name":"Conscientiousness",
            "code": 3
        },
        4: {
            "name":"Neuroticism",
            "code": 4
        },
        5: {
            "name":"Intellect/Imagination",
            "code": 5
            }
        }

def sum_personality(item, code, data):
    count = 0
    for column in data.columns:
        score = item[column]
        if type(score) == str:
            result = int(score[4])
            if int(score[1]) == catogories[code]["code"]:
                count += result
    return count
    
```

```python
for cat in catogories:

    pscores_df[catogories[cat]['name']] = pscores_df.apply(lambda row: sum_personality(row, cat, pscores_df), axis=1)
    
pscores_df.head()
```

# High Risk vs Low Risk

```python
is_high_risk = (dpt_pscores["Neuroticism"] < 30) & (dpt_pscores["Conscientiousness"] < 30) & (dpt_pscores["Agreeableness"] < 30)
high_risk = dpt_pscores[is_high_risk]
high_risk["IPIP_HIGH_RISK"] = 'high_risk'

# low_risk = dpt_pscores[]
high_risk[["Department"]]
```

```python
is_low_risk = (dpt_pscores["Neuroticism"] >= 30) & (dpt_pscores["Conscientiousness"] >= 30) & (dpt_pscores["Agreeableness"] >= 30)
low_risk = dpt_pscores[is_low_risk]
low_risk["IPIP_HIGH_RISK"] = 'low_risk'

low_risk.Department.unique()
```

```python
risks_df = {
    'risk': ['Low Risk', 'High Risk']
}

def count_risks (dataframe, department):
    num_of_people = len(dataframe[dataframe.Department == department])
    
    if num_of_people.index == None:
        num_of_people = 0

    if department in risks_df:
        risks_df[department].append(num_of_people)
    else:
        risks_df[department] = [num_of_people]
        
```

```python
for dpt in low_risk.Department.unique():
    count_risks(low_risk, dpt)
    
for dpt in high_risk.Department.unique():
    count_risks(high_risk, dpt)
    
risks_df
```
